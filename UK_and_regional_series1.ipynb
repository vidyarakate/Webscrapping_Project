```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://www.metoffice.gov.uk/research/climate/maps-and-data/uk-and-regional-series#Yearorder"
page=requests.get(url)
soup=BeautifulSoup(page.content,'lxml')
soup
```

```python
regions=soup.find(id="region").contents[3::2]
regions=[d['value'] for d in regions]
regions
```

```python
paras=soup.find(id="parameter").contents[3::2]
paras=[d['value'] for d in paras]
paras
```

```python
for reg in regions:
    for pr in paras:
        url=f"https://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/{pr}/ranked/{reg}.txt"
        page=requests.get(url)
        soup1=BeautifulSoup(page.content,'lxml')
        data=soup1.find('p').get_text()
        obj=open(f'data/{reg}_{pr}.txt','w')
        obj.write(data)
        obj.close()
        
```
