```python
import requests
from bs4 import BeautifulSoup
url='https://en.wikipedia.org/wiki/States_and_union_territories_of_India'
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
#soup
```

```python
data=soup.find('table',{"class":"wikitable sortable plainrowheaders"})
data
```

```python
heads=data.find_all('th')
heads
headings=[h.get_text().strip()for h in heads][:11]
headings
```

```python
states=[h.get_text().strip()for h in heads][11:]
states
```

```python
table_body=data.find_all('tr')[1:]
table_body
```

```python
first=[d.get_text().strip() for d in table_body[0].find_all('td')]
first
```

```python
comp=[]
for t in table_body:
    l=[d.get_text().strip()for d in t.find_all('td')]
    comp.append(l)
len(comp),comp
```

```python
for c in comp:
    print(c)
```
