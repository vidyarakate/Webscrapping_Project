## Required data from 1-12-21 to till yesterday


```python
import datetime
base=datetime.datetime.today()
t=base-datetime.datetime(2021,12,1)
date_list=[base-datetime.timedelta(days=x) for x in range(t.days+1)]
dates=[d.strftime("%d-%m-%Y") for d in date_list]
dates.reverse()
dates
```

```python
today=[base-datetime.datetime(2021,12,1)]
today
```

```python
for s in dates:
    s1=s.replace('-','/')
    url='https://www.delhisldc.org/Loaddata.aspx?mode={s1}'
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'lxml')
    data=soup.find(id='ContentPlaceHolder3_DGGridAv')
    rows=data.find_all('tr')
    heads=[p.get_text() for p in rows[0].find_all('td')]
    table=[]
    for d in rows[1:]:
        #print(d)
        table.append([p.get_text() for p in d.find_all('td')])
    dict={}
    for k in heads:
        dict[k]=[]
    for t in table:
        dict['TIMESLOT'].append(t[0])
        dict['DELHI'].append(t[1])
        dict[ 'BRPL'].append(t[2])
        dict['BYPL'].append(t[3])
        dict['NDPL'].append(t[4])
        dict['NDMC'].append(t[5])
        dict['MES'].append(t[6])
    df=pd.DataFrame(dict)
    r=s1.replace('/','_')
    df.to_csv(f'state_load_Data/{r}.csv',index=False)
    
    
        
```
