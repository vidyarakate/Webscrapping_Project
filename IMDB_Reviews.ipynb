```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.imdb.com/search/title/?release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start=1&ref_=adv_nxt'
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
#soup
```

```python
all_movies=soup.find_all(class_='lister-item mode-advanced')
all_movies
```

```python
name,rating,metas,votes,gross=([],[],[],[],[])
for first in all_movies:
    if first.find(class_='inline-block ratings-metascore')!=None:
        name.append(first.find(class_='lister-item-header').get_text().replace('\n','').replace(' ',''))
        rating.append(first.find(class_='inline-block ratings-imdb-rating').get_text().strip())
        metas.append(first.find(class_='inline-block ratings-metascore').get_text().strip('\n  Metascore '))
        
        vg=first.find(class_='sort-num_votes-visible').get_text().strip()
        
        if len(vg)==2:
            votes.append(first.find(class_='sort-num_votes-visible').get_text().strip().split('|')[0].replace('Votes:\n','').strip())
            gross.append(first.find(class_='sort-num_votes-visible').get_text().strip().split('|')[1].replace('Gross:\n','').strip())
        else:
            votes.append(first.find(class_='sort-num_votes-visible').get_text().strip().split('|')[0].replace('Votes:\n','').strip())
            gross.append(None)
            
    
```

```python
len(name),len(rating),len(metas),len(votes),len(gross)
```

```python
import pandas as pd
d={'Names':name,'Rating':rating,'Metascore':metas,'Votes':votes,'Gross':gross}
df=pd.DataFrame(d)
df
```

```python
df.to_csv("movie_ratings.csv",index=False)
```
