```python
import requests
from bs4 import BeautifulSoup
url='https://www.bookdepository.com/bestsellers'
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
all_books=soup.find_all(class_='book-item')
len(all_books)
```

```python
first=all_books[0]
#first
```

```python
bname=first.find(class_="title").get_text().strip().replace('\n','')
bname
```

```python
autname=first.find(class_='author').get_text().strip()
autname
```

```python
fpr=first.find(class_="price-wrap").get_text().strip().split('\n')[0]
fpr
```

```python
bname,autname,fpr
```

```python
Name,Author,Price=([],[],[])
for first in all_books:
    if first.find(class_='author').get_text().strip()!=None:
        Name.append(first.find(class_="title").get_text().strip().replace('\n',''))
        Author.append(first.find(class_='author').get_text().strip())
        Price.append(first.find(class_="price-wrap").get_text().strip().split('\n')[0])
```

```python
len(Name),len(Author),len(Price)
```

```python
import pandas as pd
d={'Name':Name,'Author':Author,'Price':Price}
df=pd.DataFrame(d)
df
```
